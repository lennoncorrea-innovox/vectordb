## Qdrant

```python
from db.qdrant import Qdrant
qdrant = Qdrant()
```

### Create Collections

```python
qdrant.create_collection("test_collection" , 128, "a")
```

### Add

```python
import numpy as np
vectors = np.random.random((10, 128)).astype('float32')
```

```python
qdrant.add("test_collection", vectors, payloads=[{"color": "red", "size": "big"} for i in range(10)])
qdrant.add("test_collection", vectors, payloads=[{"color": "black",  "size": "big"} for i in range(10)])
qdrant.add("test_collection", vectors, payloads=[{"color": "red",  "size": "small"} for i in range(10)])
qdrant.add("test_collection", vectors, payloads=[{"color": "black",  "size": "small"} for i in range(10)])
```

### Search

```python
results = qdrant.search("test_collection", [vectors[0], vectors[1]], filter={"color": "red"})
results_2 = qdrant.search("test_collection", [vectors[0], vectors[1]], filter={"color": "black"})
results_3 = qdrant.search("test_collection", [vectors[0], vectors[1]], filter={"color": "red", "size": "big"})

print(results)
print(results_2)
print(results_3)
```

```python
id_0 = results[0][0].id
id_1 = results[0][1].id
```

### Get

```python
qdrant.get("test_collection", id_0)
```

```python
qdrant.get("test_collection", [id_0, id_1])
```

### Remove Collection

```python
qdrant.delete_collection("test_collection")
```
